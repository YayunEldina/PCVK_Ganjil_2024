2. Apakah perbedaan dan pengaruhnya pembuatan black image antara tipe data int16
dan int32?
* Jawab : Perbedaan dan pengaruh pembuatan black image antara tipe data `int16` dan int32 terletak pada ukuran bit dan rentang nilainya.

- **int16:** Menggunakan 16-bit (2 byte) untuk setiap piksel dengan rentang nilai dari -32,768 hingga 32,767, termasuk nol. Saat digunakan untuk gambar hitam, mungkin ada perbedaan kecil pada nilai piksel, namun secara keseluruhan gambar tetap terlihat hitam.
  
- **int32:** Memakai 32-bit (4 byte) untuk setiap piksel, dengan rentang nilai yang jauh lebih besar dari -2.147.483.648 hingga 2.147.483.647, termasuk nol. Pada gambar hitam, tidak ada perbedaan visual yang tampak, tetapi penggunaan memori lebih besar dibandingkan dengan int16.

2. Tampilkan image dalam channel Red-Blue dan Green-Blue saja!

```python
# Channel Red-Blue
red_blue = img.copy()
red_blue[:, :, 1] = 0  # Set saluran hijau menjadi 0 (hanya tinggalkan saluran merah dan biru)

# Channel Green-Blue
green_blue = img.copy()
green_blue[:, :, 2] = 0  # Set saluran merah menjadi 0 (hanya tinggalkan saluran hijau dan biru)

red_blue_rgb = cv.cvtColor(red_blue, cv.COLOR_BGR2RGB)
green_blue_rgb = cv.cvtColor(green_blue, cv.COLOR_BGR2RGB)

# Menggabungkan gambar secara horizontal
final_frame = np.hstack((red_blue_rgb, green_blue_rgb))

# Menampilkan gambar yang digabungkan
plt.imshow(final_frame)
plt.axis('off')
plt.title('Red-Blue and Green-Blue Channels')
plt.show()
```

5. Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image.
disarankan menggunakan bilangan acak/random!

```python
import random

# Buat gambar kosong (hitam)
image = np.zeros(shape=(400,400,3),dtype=np.int32)

# Buat 5 kotak dengan ukuran dan warna acak
for _ in range(5):
    # Tentukan ukuran acak antara 20 dan 100
    width = random.randint(20, 100)
    height = random.randint(10, 100)

    # Tentukan posisi acak
    x = random.randint(0,400 - width)  # Maksimal x adalah 400 agar kotak tidak melewati batas gambar
    y = random.randint(0,400 - height)  # Maksimal y adalah 400 agar kotak tidak melewati batas gambar

    # Tentukan warna acak
    color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))

    # Gambar kotak dengan ukuran, posisi, dan warna yang sudah ditentukan
    cv.rectangle(image, (x, y), (x + width, y + height), color, 2)

cv2_imshow(image)
```

8. Buat rectangle pada bagian sudut bawah kiri channel B pada color space RGB
dari citra kitten/ lena/ mandrill/ male/ female/ couple/ sailboat/ peppers!

```python
# Membaca gambar
kitten = img.copy()

# Menentukan ukuran dan warna kotak (rectangle)
rect_width, rect_height = 200, 200
green_color = (0, 255, 0)  # Mengganti warna menjadi hijau

# Menambahkan kotak (rectangle) pada sudut bawah kiri di saluran hijau (Green channel)
# Pada array, slicing dilakukan untuk memilih area tertentu dari gambar.
kitten[-rect_height:, :rect_width, 1] = green_color[1]  # Menetapkan hanya nilai hijau pada saluran 1

# Menampilkan gambar yang telah dimodifikasi
cv2_imshow(kitten)
```
