```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow  # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

```python
# Membuat list untuk menyimpan url dari beberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://cdns.klimg.com/merdeka.com/i/w/news/2021/02/04/1271213/540x270/penuh-tatapan-cinta-10-gambar-kucing-lucu-ini-akan-menghangatkan-hatimu.jpg"]

# baca dan tampilkan image
# loop pada tiap url image, beberapa imae dapat disimpan pada list

for url in urls:
  image = io.imread(url)                              #read image
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5)     #resize image to half size
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)      #convert color to RGB
  final_frame = cv.hconcat((image, image_2))          #concatenate image
  cv2_imshow(final_frame)                             #show image
  print('\n')
```

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ", tinggi," x ", lebar)
cv2_imshow(image_2)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)


# Membuat garis horizontal di tengah image
for y in range(lebar):
    image_3[int(tinggi/2), y] = [255, 255, 255]

# Menggabungkan gambar sebelum dan sesudah penambahan garis
final_frame = cv.hconcat([image_2, image_3])

# Menampilkan gambar
cv2_imshow(final_frame)
```

4. Buat kotak menggunakan kumpulan pixel warna putih di sembarang tempat dalam gambar

```python
import cv2 as cv
from google.colab.patches import cv2_imshow

# Mengubah gambar dari BGR ke RGB
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = image_2.copy()  # Salin gambar RGB untuk memodifikasi

# Tentukan ukuran kotak (lebar dan tinggi)
kotak_lebar = int(lebar / 3)
kotak_tinggi = int(tinggi / 3)  # Menetapkan tinggi kotak (1/3 dari tinggi gambar)

# Atur offset untuk menggerakkan kotak
offset_x = -0  # Posisi horizontal, positif ke kanan, negatif ke kiri
offset_y = 20  # Posisi vertikal, positif ke bawah, negatif ke atas

# Tentukan posisi untuk menempatkan kotak dengan offset
start_x = int((lebar - kotak_lebar) / 2) + offset_x  # Posisi horizontal dengan offset
start_y = int((tinggi - kotak_tinggi) / 2) + offset_y  # Posisi vertikal dengan offset

# Membuat kotak dengan warna putih pada lokasi yang dipilih
for x in range(start_x, start_x + kotak_lebar):
    for y in range(start_y, start_y + kotak_tinggi):
        if 0 <= x < lebar and 0 <= y < tinggi:  # Pastikan posisi berada dalam batas gambar
            image_3[y, x] = [255, 255, 255]  # Mengubah piksel menjadi putih

# Menggabungkan gambar asli dengan gambar yang telah dimodifikasi
final_frame = cv.hconcat([image_2, image_3])

# Menampilkan gambar
cv2_imshow(final_frame)
```

3. Tutup tiap bagian tertentu dari kartu identitas tersebut. Kelompok 4 tutup bagian Alamat dan Kecamatan. Tutup menggunakan 2 kotak warna yang berbeda. Tidak boleh menggunakan fungsi rectangle openCV.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import urllib.request

# Langkah 1: Unduh gambar dari URL
url = "https://bimamedia-gurusiana.ap-south-1.linodeobjects.com/099fe6b0b444c23836c4a5d07346082b/2021/04/20/l-img20210420015823jpg20210420005933.jpeg"
resp = urllib.request.urlopen(url)
image = np.asarray(bytearray(resp.read()), dtype="uint8")
image = cv2.imdecode(image, cv2.IMREAD_COLOR)

# Langkah 2: Konversi gambar dari BGR ke RGB untuk Matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Langkah 4: Tentukan offset
offset_x = -10   # Offset horizontal, positif untuk bergerak ke kanan, negatif untuk bergerak ke kiri
offset_y = -70   # Offset vertikal, positif untuk bergerak ke bawah, negatif untuk bergerak ke atas

# Langkah 5: Sesuaikan offset untuk kotak hijau dan merah
# Naikkan offset vertikal untuk kotak merah agar kotak tersebut naik
red_box_offset_y = offset_y - 30  # Geser kotak merah ke atas

# Turunkan offset vertikal untuk kotak hijau agar kotak tersebut turun
green_box_offset_y = offset_y + 70  # Geser kotak hijau ke bawah

# Koordinat untuk kotak 'Alamat' dengan offset yang diperbarui
alamat_top_left = (200 + offset_x, 280 + red_box_offset_y)
alamat_bottom_right = (400 + offset_x, 310 + red_box_offset_y)

# Warna kotak merah sesuai tampilan asli (RGB: Merah)
color_red = [255, 0, 0]  # Warna merah dalam format RGB

# Tutup 'Alamat' dengan kotak merah
image_rgb[alamat_top_left[1]:alamat_bottom_right[1], alamat_top_left[0]:alamat_bottom_right[0]] = color_red

# Koordinat untuk kotak 'Kecamatan' dengan offset yang diperbarui
kecamatan_top_left = (200 + offset_x, 250 + green_box_offset_y)
kecamatan_bottom_right = (400 + offset_x, 280 + green_box_offset_y)

# Warna kotak hijau sesuai tampilan asli (RGB: Hijau)
color_green = [0, 255, 0]  # Warna hijau dalam format RGB

# Tutup 'Kecamatan' dengan kotak hijau
image_rgb[kecamatan_top_left[1]:kecamatan_bottom_right[1], kecamatan_top_left[0]:kecamatan_bottom_right[0]] = color_green

# Konversi gambar yang dimodifikasi dari RGB ke BGR untuk OpenCV
image_rgb_mod = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2BGR)

# Menggabungkan gambar asli dengan gambar yang telah dimodifikasi secara horizontal
final_frame = cv2.hconcat([image, image_rgb_mod])

# Menampilkan gambar hasil gabungan
plt.imshow(cv2.cvtColor(final_frame, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
